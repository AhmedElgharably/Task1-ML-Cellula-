```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style("whitegrid")
plt.rcParams["figure.figsize"]=(8,5)
```

```python
df = pd.read_csv(r"C:\Users\Ahmed Elghrably\OneDrive\Desktop\final_internship_data.csv")
```

```python
df.dropna(inplace=True)
```

```python
df.isnull().sum()
```

```python
df.drop(columns=[
    "User ID",
    "User Name",
    "Driver Name",
    "key"
], inplace=True)
```

```python
df["pickup_datetime"] = pd.to_datetime(df["pickup_datetime"])
```

```python

###Q1. Does higher passenger count lead to higher fares

sns.scatterplot(
    data=df,
    x="passenger_count",
    y="fare_amount"
)
plt.title("Passenger Count vs Fare Amount")
plt.show()
```

```python
#Q2. Does car condition influence fare
sns.boxplot(
    data=df,
    x="Car Condition",
    y="fare_amount"
)
```

```python
#Q3. At what hour are fares highest
hourly = df.groupby("hour")["fare_amount"].mean()

sns.lineplot(
    x=hourly.index,
    y=hourly.values,
    marker="o"
)
```

```python
#Q4. Does traffic condition affect fare
sns.boxplot(
    data=df,
    x="Traffic Condition",
    y="fare_amount"
)
```

```python
#Q5. Is distance the strongest predictor
sns.scatterplot(
    data=df,
    x="distance",
    y="fare_amount"
)
```

```python
Q6. When are ride requests most frequent
sns.countplot(
    data=df,
    x="hour"
)
```

```python
#Q7. Does weather affect trip distance
sns.boxplot(
    data=df,
    x="Weather",
    y="distance"
)
```

```python
#Q8. Are rides near airports more expensive
sns.scatterplot(
    data=df,
    x="jfk_dist",
    y="fare_amount"
)
```

```python
#How is fare amount distributed? (Histogram
plt.figure(figsize=(8,5))

sns.histplot(df["fare_amount"], bins=40, kde=True)

plt.title("Distribution of Fare Amount")
plt.xlabel("Fare Amount ($)")
plt.ylabel("Frequency")

plt.show()
```

```python
#Are there outliers in fare? (Boxplot)
plt.figure(figsize=(8,2))

sns.boxplot(x=df["fare_amount"])

plt.title("Outliers in Fare Amount")

plt.show()
```

```python
#Which weather condition is most common? (Countplot)
plt.figure(figsize=(7,5))

sns.countplot(data=df,
              x="Weather",
              order=df["Weather"].value_counts().index)

plt.title("Weather Condition Distribution")

plt.show()
```

```python
#Which traffic condition is most common? (Countplot)
plt.figure(figsize=(7,5))

sns.countplot(data=df,
              x="Traffic Condition",
              order=df["Traffic Condition"].value_counts().index)

plt.title("Traffic Condition Distribution")

plt.show()
```

```python
#Does weather affect fare
plt.figure(figsize=(8,5))

sns.boxplot(data=df,
            x="Weather",
            y="fare_amount")

plt.title("Fare Amount by Weather")

plt.show()
```

```python
#Does weekday affect fare?
weekday_avg = df.groupby("weekday")["fare_amount"].mean().reset_index()

plt.figure(figsize=(8,5))

sns.barplot(data=weekday_avg,
            x="weekday",
            y="fare_amount")

plt.title("Average Fare by Weekday")

plt.xlabel("Weekday")
plt.ylabel("Average Fare")

plt.show()
```

```python
#Which month has the highest average fare?
monthly = df.groupby("month")["fare_amount"].mean()

plt.figure(figsize=(8,5))

sns.lineplot(x=monthly.index,
             y=monthly.values,
             marker="o")

plt.title("Average Fare by Month")

plt.xlabel("Month")
plt.ylabel("Average Fare")

plt.show()
```

```python
#Correlation Heatmap
plt.figure(figsize=(12,10))

corr = df.select_dtypes(include="number").corr()

sns.heatmap(corr,
            annot=True,
            cmap="coolwarm",
            fmt=".2f")

plt.title("Correlation Matrix")

plt.show()
```

```python
#pair plot
cols = [
    "fare_amount",
    "distance",
    "passenger_count",
    "hour",
    "jfk_dist"
]

sns.pairplot(df[cols])

plt.show()
```
